```python
import os
import sys
import time
from io import StringIO
from google import genai
from google.genai import errors
from kaggle_secrets import UserSecretsClient
import gradio as gr

# --- Configuration & Templates ---
user_age_group = "Junior"

PROMPT_TEMPLATES = {
    "Junior": """
    You are a whimsical, patient, and encouraging AI Storytelling Companion for young children (ages 6-10).
    Your job is to co-write a story step-by-step with the child.
    RULES:
    - Use clear, simple language at a 3rd-grade reading level.
    - Write short paragraphs (maximum 2-3 sentences) so the child doesn't get overwhelmed.
    - End your turn by asking the child what happens next with an exciting, easy-to-answer question.
    - Keep themes gentle and fun (e.g., magical animals, space adventures, friendly superheroes).
    """,
    "Advanced": """
    You are a sophisticated Creative Writing Mentor and Editor for older students and teens.
    Your job is to collaborate on an intricate, high-quality narrative.
    RULES:
    - Write with rich vocabulary, compelling pacing, and engaging descriptions.
    - Actively introduce nuanced narrative choices, environmental tension, or potential plot twists.
    - Keep your responses to 1-2 paragraph narrative blocks, then pass the turn back to the user.
    - Allow for diverse genres like Mystery, Science Fiction, or Epic Fantasy.
    """
}

# --- Robust Multi-Agent Engine ---
class StableStorytellingSystem:
    def __init__(self, age_group):
        self.age_group = age_group
        self.system_instruction = PROMPT_TEMPLATES[age_group]
        self.story_history = []
        
        user_secrets = UserSecretsClient()
        secret_key = user_secrets.get_secret("GEMINI_API_KEY")
        self.client = genai.Client(api_key=secret_key)
        
    def play_turn(self, user_text):
        coach_feedback = ""
        if self.age_group == "Junior":
            if len(user_text.split()) < 3:
                coach_feedback = "💡 Coach Tip: Try adding a few more descriptive words to expand your world!"
            elif not user_text.endswith(('.', '!', '?')):
                coach_feedback = "💡 Coach Tip: Brilliant sentence! Remember to place a punctuation mark at the very end."
            else:
                coach_feedback = "💡 Coach Tip: Spot on! Your grammar is looking excellent."
        else:
            if "and then" in user_text.lower():
                coach_feedback = "💡 Editor Notes: Try swapping out 'and then' with a transitional marker."
            else:
                coach_feedback = "💡 Editor Notes: Excellent phrase configuration."

        self.story_history.append({"role": "user", "parts": [user_text]})
        full_contents = [self.system_instruction] + [
            f"{'User' if h['role']=='user' else 'AI'}: {h['parts'][0]}" for h in self.story_history
        ]
        
        # --- SAFE GENERATION WITH AUTOMATIC ERROR CATCHING ---
        ai_response = ""
        try:
            # Using standard naming convention supported by the unified SDK
            response = self.client.models.generate_content(
                model="gemini-2.5-flash", 
                contents="\n".join(full_contents)
            )
            ai_response = response.text.strip()
        except Exception as e:
            # If a 503 or 429 error occurs, catch it and return a friendly, graceful message instead of breaking the UI
            ai_response = "*(The winter wind is blowing hard outside and Google's free-tier servers are temporarily busy! Please wait a few seconds and try sending your message again.)*"

        self.story_history.append({"role": "model", "parts": [ai_response]})
        
        print(f"{coach_feedback}")
        print(f"📌 NARRATIVE UPDATE:")
        print(f"{ai_response}")

# Reset backend tracking engine instance
engine = StableStorytellingSystem(user_age_group)

# --- Gradio Interface ---
def chatbot_response(user_message, history):
    if isinstance(user_message, dict) and "text" in user_message:
        actual_text = user_message["text"]
    else:
        actual_text = str(user_message)
        
    old_stdout = sys.stdout
    sys.stdout = mystdout = StringIO()
    
    engine.play_turn(actual_text)
    
    sys.stdout = old_stdout
    return mystdout.getvalue()

forced_light_css = """
.gradio-container {
    background-color: #eef5fc !important;
    background-image: url('https://www.transparenttextures.com/patterns/snow.png') !important;
    background-repeat: repeat !important;
}
.chatbot, .chatbot .message-wrap, .chatbot .message {
    background-color: #ffffff !important;
}
.chatbot .bot .message-text, .chatbot .bot p, .chatbot .bot span,
.chatbot .user .message-text, .chatbot .user p, .chatbot .user span,
.gradio-container h1, .gradio-container p {
    color: #0b3c5d !important;
    font-weight: bold !important;
}
.chatbot .user {
    background-color: #bce0fd !important;
    border-radius: 8px !important;
}
"""

with gr.Blocks(css=forced_light_css) as demo:
    gr.Markdown("# ❄️ AI Interactive Winter Storytelling & Literacy Assistant")
    gr.Markdown(f"Currently running in **[{user_age_group}]** Mode. Let's write a magical story!")
    
    gr.ChatInterface(
        fn=chatbot_response,
        type="messages"
    )

demo.launch(inline=True, quiet=True)
```
